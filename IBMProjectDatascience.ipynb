##**4.1) Profit and Sales analysis based on Sub-Category**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plot line chart for category and subcategory sales
sale_df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()

# Plot settings
plt.title('Total Profit and Sales according to Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Profit and Sales')
plt.rcParams['figure.figsize'] = [10 ,6]
# Show the plot
plt.show()
```

In the above plot, we can see the comparison of Sales and profit in Sub-Category together. It makes it easier to identify which has made most and least sales and where profit is made at a greater range.

**4.2)  Sales and Profit Comparison by Ship Mode and Region**

```python
import matplotlib.pyplot as plt

# Group the DataFrame by Ship Mode, Region, and calculate the total sales and profit
sales_profit_by_ship_mode_region = sale_df.groupby(['Ship Mode', 'Region'])['Sales', 'Profit'].sum().reset_index()

# Pivot the DataFrame to make Region values as columns
pivot_table = sales_profit_by_ship_mode_region.pivot(index='Ship Mode', columns='Region', values=['Sales', 'Profit'])

# Plot the bar chart
pivot_table.plot.bar()

# Set the x-axis label, y-axis label, and title
plt.xlabel('Ship Mode')
plt.ylabel('Amount')
plt.title('Sales and Profit Comparison by Ship Mode and Region')

# Display the plot
plt.show()
```

In the above plot, we can see that 4 parameters are used, which are sales and profit based on Ship Mode and Region are compared.

###**4.5) Line Plot of Discount VS Profit**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a DataFrame named "sale_df" with columns "Discount" and "Profit"

# Plotting the line plot
plt.figure(figsize=(8, 6))
sns.lineplot(x='Discount', y='Profit', data=sale_df, color='green', label='Discount vs. Profit')

# Adding labels and title
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.title('Profit vs. Discount')
plt.legend()

# Display the plot
plt.show()



```
